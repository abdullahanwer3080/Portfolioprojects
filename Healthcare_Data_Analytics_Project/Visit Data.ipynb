```python
def department_wait_time(df,department):
  department_wait_time_hrs=df[department]/60
  department_max_wait_time=department_wait_time_hrs.max()
  department_min_wait_time=department_wait_time_hrs.min()
  department_avg_wait_time=department_wait_time_hrs.mean()
  department_std_wait_time=department_wait_time_hrs.std()
  return f"department_max_wait_time: {department_max_wait_time},department_min_wait_time: {department_min_wait_time},department_avg_wait_time: {department_avg_wait_time} ,department_std_wait_time: {department_std_wait_time}"
```

```python
department_wait_time(df,'Registration_minutes')
```

```python
df2=df[mask].copy()
df3=df[mask2].copy()

df2['Visit_Date']=pd.to_datetime(df2['Visit_Date'])
df3['Visit_Date']=pd.to_datetime(df3['Visit_Date'])

df2.set_index('Visit_Date',inplace=True)
df3.set_index('Visit_Date',inplace=True)

df4_resample=df2['Total_Cost'].resample('YE').sum()
df5_resample=df3['Total_Cost'].resample('YE').sum()

plt.figure(figsize=(10,6))
plt.plot(df4_resample.index,df4_resample.values, label='With Insurance')
plt.plot(df5_resample.index,df5_resample.values, label='Without Insurance')
plt.xlabel('Year')
plt.ylabel('Total Cost')
plt.title('Yearly Total Cost by Insurance Status')
plt.legend()
plt.show()
```

```python
import folium
```

```python
m=folium.Map(location=[df['Latitude'].iloc[0],df['Longitude'].iloc[0]])

```

```python
for index,row in df.iterrows():
    folium.Marker(location=[row['Latitude'],row['Longitude']],popup=row['Facility Type']).add_to(m)
m
```

```python
from folium.plugins import MarkerCluster
```

```python
marker_cluster=MarkerCluster()
```

```python
for index,row in df.iterrows():
    marker=folium.Marker(location=[row['Latitude'],row['Longitude']],popup=row['Facility Type'])
    marker.add_to(marker_cluster)

marker_cluster.add_to(m)

m
```
